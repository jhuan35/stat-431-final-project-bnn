```python
import pandas as pd
import yfinance as yf
```

```python
tickerSymbol = 'SPY'
period='1d'
start='2010-1-1'
end='2023-1-1'
```

```python
tickerData = yf.Ticker(tickerSymbol)

tickerDf = tickerData.history(period=period, start=start, end=end)

tickerDf.head(20)
```

```python
tickerDf.to_csv('SPY_historical_data.csv', index=True)
```
